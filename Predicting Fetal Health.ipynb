```python
# Import pandas
import pandas as pd

# Read in dataset
# ...WRITE YOUR CODE FOR TASK 1 ...
df = pd.read_csv("/content/drive/MyDrive/datasets and projects/fetal_health.csv")

# Inspect data
df
```

```python
duplicates = df.duplicated().sum()
# Inspect data
duplicates
```

```python
df = df.drop_duplicates()


df
```

```python
null_values = df.isnull().sum()
# Inspect data
null_values
```

```python
values = df["fetal_health"].value_counts()
# Inspect data
values
```

```python
# Split into X and Y
x = df.drop("fetal_health", axis=1) #.reset_index(drop=True)
y = df["fetal_health"] #.copy() #.reset_index(drop=True)

x.shape,
y.shape
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
# Inspect data
x_train, x_test, y_train, y_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model1b  = LogisticRegression(solver='liblinear',
                         penalty='l2',
                         max_iter=100)
model1b.fit(x_train, y_train)
```

```python
y_predict1b = model1b.predict(x_test)
y_predict1b
```

```python
accuracy_lr = model1b.score(x_test, y_test)
accuracy_lr
```

```python
from sklearn.ensemble import RandomForestClassifier
model2b = RandomForestClassifier(criterion='gini', n_estimators=100, max_depth=4, random_state=33)
```

```python
model2b = model2b.fit(x_train, y_train)
```

```python
y_pred2b = model2b.predict(x_test)
y_pred2b
```

```python
accuracy_rf = accuracy_score(y_test, y_pred2b)
accuracy_rf
```

```python
from sklearn.tree import DecisionTreeClassifier
model3b = DecisionTreeClassifier(criterion='gini', max_depth=2, random_state=33)
```

```python
model3b = model3b.fit(x_train, y_train)
```

```python
y_pred3b = model3b.predict(x_test)
y_pred3b
```

```python
accuracy_dt = accuracy_score(y_test, y_pred3b)
accuracy_dt
```
